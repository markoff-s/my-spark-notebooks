```python
import pprint

from operator import add
from pyspark.sql import SparkSession


spark = SparkSession\
        .builder\
        .appName("PythonWordCount")\
        .getOrCreate()

file_path = "/tools/spark-2.4.5-bin-hadoop2.7/examples/src/main/python/wordcount.py"
lines = spark.read.text(file_path).rdd.map(lambda r: r[0])

counts = lines.flatMap(lambda x: x.split()) \
                    .map(lambda x: (x, 1)) \
                    .reduceByKey(add)

output = counts.collect()
for (word, count) in output:
    print(f"{word} = {count}")

#pprint.pprint(counts.collect())
#pprint.pprint(lines.collect())

spark.stop()
```
